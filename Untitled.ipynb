```python
import numpy as np
from keras.models import Model
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Activation
from keras.layers import Masking
from keras.optimizers import RMSprop
from keras import backend as k
from sklearn.preprocessing import normalize

def load_file(name):
    with open(name, 'r') as file:
        return np.loadtxt(file, delimiter=',')

train = load_file('train.csv')
test_x = load_file('test_x.csv')
test_y = load_file('test_y.csv')

#train[:, 2:26] = normalize(train[:, 2:26], axis=0)
#test_x[:, 2:26] = normalize(test_x[:, 2:26], axis=0)

# Make engine numbers and days zero-indexed, for everybody's sanity
train[:, 0:2] -= 1
test_x[:, 0:2] -= 1

#max look-back time
max_time = 100

def build_data(engine, time, x, max_time):
    
    # y[0] is days remaining
    out_y = np.empty((0), dtype=np.float32)
    out_x = np.empty((0, max_time, 24), dtype=np.float32)

    for i in range(100):#100 engine
        #print("Engine: " + str(i))
        
        max_engine_time = int(np.max(time[engine == i])) + 1

        start = max_engine_time - 1

        this_x = np.empty((0, max_time, 24), dtype=np.float32)

        for j in range(start, max_engine_time):
            engine_x = x[engine == i]
            xtemp = np.zeros((1, max_time, 24))
            xtemp[:, max_time-min(j, 99)-1:max_time, :] = engine_x[max(0, j-max_time+1):j+1, :]
            this_x = np.concatenate((this_x, xtemp))
            
        out_y = np.append(out_y, max_engine_time)
        out_x = np.concatenate((out_x, this_x))

    return out_x, out_y

train_x, train_y = build_data(train[:, 0], train[:, 1], train[:, 2:26], max_time)
test_x, test_y = build_data(test_x[:, 0], test_x[:, 1], test_x[:, 2:26], max_time)

print(train_x[0])
```

```python

model = Sequential()
model.add(Masking(mask_value=0., input_shape=(max_time, 24)))
model.add(LSTM(100, input_dim=100, input_length=24))
model.add(Dense(1))
model.compile(loss='mse', optimizer='adam')

# Fit!
#model.fit(train_x, train_y, nb_epoch=250, batch_size=2000, verbose=2, validation_data=(test_x, test_y))
history_object = model.fit(train_x, train_y, nb_epoch=20, batch_size=2048, verbose=2, validation_data=(test_x, test_y))

test_predict = model.predict(train_x)

print(test_predict)
print(test_y)

plt.plot(history_object.history['loss'])
plt.plot(history_object.history['val_loss'])
plt.title('loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training set', 'validation set'], loc='upper right')
plt.show()
```
